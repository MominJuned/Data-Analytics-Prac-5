```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### Generate Dataset (Data Collection Phase)

```python
np.random.seed(40)

data = {
    "Advertising": np.random.randint(2000,10000,400),
    "Price": np.random.randint(100,1000,400),
    "Discount": np.random.randint(5,40,400),
    "Sales": np.random.randint(20000,80000,400)
}

df = pd.DataFrame(data)

df.head()
```

```python
# Phase 1 — Business Understanding
```

```python
# # Business problem:

# A company wants to understand how advertising, price, and discount affect sales in order to increase revenue.
```

```python
# Goal:

# Predict sales and identify important factors.
```

```python
# Phase 2 — Data Understanding
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Visualization — Data Understanding
```

```python
# Sales Distribution

plt.hist(df["Sales"], bins=25,color="g")
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Graph — Boxplot

plt.boxplot(df["Sales"])
plt.title("Sales Outlier Detection")
plt.show()
```

```python
# Phase 3 — Data Preparation
```

```python
X = df[["Advertising","Price","Discount"]]
y = df["Sales"]
```

```python
X.shape
```

```python
# Phase 4 — Modeling
```

```python
# Apply Linear Regression Model.
```

```python
model = LinearRegression()

model.fit(X,y)
```

```python
# Model coefficients.
model.coef_
```

```python
# Intercept.
model.intercept_
```

```python
# Phase 5 — Prediction
```

```python
# Predict sales for new marketing strategy.

# Example:

# Advertising = 7000
# Price = 500
# Discount = 20
```

```python
model.predict([[7000,500,20]])
```

```python
# Visualization — Relationship Analysis
```

```python
plt.scatter(df["Advertising"], df["Sales"])
plt.title("Advertising vs Sales")
plt.xlabel("Advertising Budget")
plt.ylabel("Sales")
plt.show()
```

```python
# Insight:

# Higher advertising budgets generally increase sales.
```

```python
# Phase 6 — Evaluation
```

```python
# Check model accuracy using R² score.
```

```python
model.score(X,y)

# Higher value → better prediction.
```
